```python
# Customer Segmentation using K-Means

# 1. Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# 2. Load Datase
df = pd.read_csv(r"C:\Users\Moorthy\Downloads\Mall_Customers.csv")
df.head()

# 3. Data Info
print(df.info())
print(df.describe())
print(df.isnull().sum())

# 4. Select features for clustering
X = df[['Annual Income (k$)', 'Spending Score (1-100)']]

# 5. Standardize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 6. Elbow Method to determine optimal K
inertia = []
K_range = range(1, 11)
for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)

plt.figure(figsize=(8, 5))
plt.plot(K_range, inertia, 'bo-')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Inertia')
plt.title('Elbow Method For Optimal K')
plt.grid(True)
plt.show()

# 7. Apply KMeans with optimal K (say, K=5)
kmeans = KMeans(n_clusters=5, random_state=42)
df['Cluster'] = kmeans.fit_predict(X_scaled)

# 8. Visualize Clusters
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Annual Income (k$)', y='Spending Score (1-100)',
                hue='Cluster', palette='Set2', data=df, s=100)
plt.title('Customer Segments (K-Means Clustering)')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend(title='Cluster')
plt.show()
```
